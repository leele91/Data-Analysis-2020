### address3.csv 를 읽어서 지도위에 표시하기

```python
df = pd.read_csv('address3.csv')
df
```

```python
df['icon'] = ['barcode', 'cloud','star','info-sign', 'headphones']
df
```

```python
# 날씨에 따른 마커 색깔
color_dict = {'Clear': 'blue', 'Snow':'white','Rain': 'gray','Extreme':'red', 'Clouds':'orange','Mist':'green'}
```

```python
mapping = folium.Map(location=[df.lat.mean(),df.lng.mean()],zoom_start=13)
mapping
```

```python
for i in df.index:
    folium.Marker(
    location=[df.lat[i], df.lng[i]],
    popup=df.bldg[i],
    tooltip=f'{df.desc[i]}, {df.temp[i]}',
    icon=folium.Icon(color=color_dict[df.weather[i]], icon=df.icon[i])
).add_to(mapping)
mapping
```

```python
df2 = pd.read_csv('상권서울.csv', encoding='utf8', sep='|')
df2.head(3)
```

```python
cafe = df2[['상호명','지점명','상권업종대분류명','상권업종중분류명','건물명','도로명주소','경도','위도']]
cafe.head()
```

```python
coffe = cafe[(cafe['상권업종중분류명']=='커피점/카페')]
coffe
```

```python
# 특정 문자열을 포함하는 데이터 필터링 : .str.contains()
# country 컬럼을 선택합니다.
# 컬럼의 값에 Korea 또는(|) Japan이라는 문자열이 포함되어있는지 판단합니다.
# 그 결과를 새로운 변수에 할당합니다.
coffe1 = coffe['도로명주소'].str.contains('서울특별시 강서구')
# 조건를 충족하는 데이터를 필터링하여 새로운 변수에 저장합니다.
coffe2 = coffe[coffe1]
coffe2
```

```python
# 스타벅스 지점 데이터
cf_stbu = coffe2[coffe2['상호명'].str.contains('스타벅스|STAR')]
cf_stbu.head(6)
```

```python
cf_stbu.groupby('상호명').count()
```

```python
cf_eda = coffe2[coffe2['상호명'].str.contains('이디야')]
cf_eda.head(6)
```

```python
cf_pabang = coffe2[coffe2['상호명'].str.contains('빽다방')]
cf_pabang.head(6)
```

```python
cf_twosome = coffe2[coffe2['상호명'].str.contains('투썸플레이스')]
cf_twosome.head(6)
```

```python
mapping1 = folium.Map(location=[37.550966,126.849532],zoom_start=13)
for i in cf_stbu.index: # 스타벅스
    folium.Marker(
        location = [cf_stbu.위도[i], cf_stbu.경도[i]],
        popup = cf_stbu.상호명[i],
        tooltip = f'{cf_stbu.상호명[i]}, {cf_stbu.지점명[i]}',
        icon=folium.Icon(color='blue')
    ).add_to(mapping1)

for k in cf_eda.index: # 이디아
    folium.Marker(
        location = [cf_eda.위도[k], cf_eda.경도[k]],
        popup = cf_eda.상호명[k],
        tooltip = f'{cf_eda.상호명[k]}, {cf_eda.지점명[k]}',
        icon=folium.Icon(color='green')
    ).add_to(mapping1)

for a in cf_pabang.index: # 빽다방
    folium.Marker(
        location = [cf_pabang.위도[a], cf_pabang.경도[a]],
        popup = cf_pabang.상호명[a],
        tooltip = f'{cf_pabang.상호명[a]}, {cf_pabang.지점명[a]}',
        icon=folium.Icon(color='magenta')
    ).add_to(mapping1)

for q in cf_twosome.index: # 투썸
    folium.Marker(
        location = [cf_twosome.위도[q], cf_twosome.경도[q]],
        popup = cf_twosome.상호명[q],
        tooltip = f'{cf_twosome.상호명[q]}, {cf_twosome.지점명[q]}',
        icon=folium.Icon(color='orange')
    ).add_to(mapping1)

mapping1
```
