## 데이터 시각화

### 시각화 패키지 맷플롯리브 소개

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

### 라인 플롯((line plot))

```python
plt.plot([1, 4, 9, 16]) # plot: 라인플롯 사용 명령
#y축을 넣으면 기본적으로 x축은 x[0, 1, 2, 3]으로 자동으로 들어감 :  / y [1, 4, 9, 16]
plt.title('line plot')
plt.show()
```

```python
plt.plot([1, 4, 9, 16]) 
plt.title('line plot')
plt.grid(True) # 그리드
plt.show()
```

```python
plt.plot([1, 4, 9, 16]) 
plt.title('line plot')
plt.grid(True)
plt.xlabel('X Axis') # x축 제목
plt.ylabel('Y Axis') # y축 제목
plt.show()
```

```python
plt.title("x ticks")
plt.plot([10, 20, 30, 40], [1, 4, 9, 16]) # x,y /x축 값을 줄려면 y축 값 앞에 적어야함
plt.grid(True)
plt.show()
```

```python
plt.title("X 틱스")# 한글이 깨짐므로 한글 폰트 사용을 설정해야함(아래 참고)
plt.plot([10, 20, 30, 40], [1, 4, 9, 16]) #x,y /x축 값을 줄려면 y축 값 앞에 적어야함
plt.grid(True)
plt.show()
```

### 한글폰트 사용

```python
import matplotlib.font_manager

matplotlib.font_manager._rebuild()
sorted([f.name for f in matplotlib.font_manager.fontManager.ttflist if f.name.startswith("Malgun")]) # 사용할 폰트 검색(맑음폰트 검색)
```

```python
# 폰트 설정
mpl.rc('font', family='Malgun Gothic') #family에 사용할 폰트를 넣음
# 유니코드에서  음수 부호설정
mpl.rc('axes', unicode_minus=False)
```

```python
plt.title('한글 제목')
plt.plot([10, 20, 30, 40], [1, 4, 9, 16])
plt.xlabel("엑스축 라벨")
plt.ylabel("와이축 라벨")
plt.show()
```

```python
font1 = {'family': 'Malgun Gothic', 'size': 24, 
         'color':  'black'}
font2 = {'family': 'Malgun Gothic', 'size': 18, 'weight': 'bold',
         'color':  'darkred'}
font3 = {'family': 'Malgun Gothic', 'size': 12, 'weight': 'light',
         'color':  'blue'}

plt.plot([10, 20, 30, 40], [1, 4, 9, 16])
plt.title('한글 제목', fontdict=font1)
plt.xlabel('엑스 축', fontdict=font2)
plt.ylabel('와이 축', fontdict=font3)
plt.show()
```
