```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
mpl.rc('font', family='Malgun Gothic') #family에 사용할 폰트를 넣음
# 유니코드에서  음수 부호설정
mpl.rc('axes', unicode_minus=False)
```

```python
X = np.linspace(0, 2*np.pi, 256)
S = np.sin(X)
C = np.cos(X)
plt.title("sine, Cosine 그래프")
plt.plot(X, S, label="sine")
plt.plot(X, C, label="cosine")
plt.grid(True)
# xticks: 눈금표시 /
# 아래처럼 tick을 표시할 값을 먼저 넣어주고, 그 다음에 그 위치에 들어갈 값을 넣어줍니다. 
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],[r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$']) # LaTeX
plt.yticks([-1, 0, 1],["Low", "Zero", "High"])
plt.legend(loc='best') # loc = 0
plt.show()
```
