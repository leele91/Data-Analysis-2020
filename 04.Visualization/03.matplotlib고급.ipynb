### Axes 객체와 subplot 명령

```python
# 2개의 플롯
x1 = np.linspace(0.0, 5.0, 100) # 노트 참고
x2 = np.linspace(0.0, 2.0, 100)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

ax1 = plt.subplot(2, 1, 1) #ax1 그래프에 행 2, 열1, 1번째 그래프
plt.plot(x1, y1, 'yo-') 
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')

ax2 = plt.subplot(2, 1, 2) #ax1 그래프에 행 2, 열1, 2번째 그래프
plt.plot(x2, y2, 'r.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.tight_layout() #
plt.show()
```

```python
# 4개의 플롯
np.random.seed(2020)

plt.subplot(221) #
plt.plot(np.random.rand(5))
plt.title("axes 1")

plt.subplot(222)
plt.plot(np.random.rand(5))
plt.title("axes 2")

plt.subplot(223)
plt.plot(np.random.rand(5))
plt.title("axes 3")

plt.subplot(224)
plt.plot(np.random.rand(5))
plt.title("axes 4")

plt.tight_layout()
plt.show()
```

```python
# 위에 있는 것을 반복문
fig, axes = plt.subplots(2, 2) # 항상 위에 이내용을 써야함

np.random.seed(0)
axes[0, 0].plot(np.random.rand(5))
axes[0, 0].set_title("axes 1")
axes[0, 1].plot(np.random.rand(5))
axes[0, 1].set_title("axes 2")
axes[1, 0].plot(np.random.rand(5))
axes[1, 0].set_title("axes 3")
axes[1, 1].plot(np.random.rand(5))
axes[1, 1].set_title("axes 4")

plt.tight_layout()
plt.show()
```

```python
# for 반복문을 사용한 subplots
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True) 
# 옵션을 줄수 있음 sharex 옆에 있는 그래프의 x축을 공유 

np.random.seed(0)
for i in range(2):
    for k in range(2):
        ax = axes[i][k]
        ax.plot(np.random.rand(5))
        ax.set_title(f"axes {2*i+k+1}")

plt.tight_layout()
plt.show()
```

```python
# for 반복문
x = np.linspace(0.0, 5.0, 100)

fig, axes = plt.subplots(2, 1, sharex=True)
for i in range(2):
    ax= axes[i]
    y = np.cos(2 * np.pi * x1) * np.exp(-x1) if i==0 else np.cos(2 * np.pi * x2)
    ylabel = 'Damped oscillation' if i == 0 else 'Undamped'
    ax.plot(x, y)
    ax.axhline(0)
    ax.set_ylim(-1, 1)
    ax.set_ylabel(ylabel)
    # ax.set_xlabel('time (s)')

plt.xlabel('time (s)')
plt.title('A tale of 2 subplots')
plt.show()
```

```python
import seaborn as sns
iris = sns.load_dataset('iris')
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 8), sharey=True)
for i, spec in enumerate(['setosa', 'versicolor', 'virginica']):
    x = iris[iris.species == spec].sepal_length
    y = iris[iris.species == spec].sepal_width
    z = iris[iris.species == spec].petal_length
    a = iris[iris.species == spec].petal_width
    ax = axes[i]
    ax.boxplot([x, y, z, a],
        labels = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
    ax.set_title(f"{spec} 품종")

plt.show()
```

### 강의 자료 - 08. 여러개의 그래프 그리는 방법

```python
ans = sns.load_dataset('anscombe')
ans.shape
```

```python
ans.head()
```

```python
ans.groupby('dataset').describe() # describe 컬럼별 요약 통계량
```

```python
ds1 = ans[ans.dataset == 'I']
ds2 = ans[ans.dataset == 'II']
ds3 = ans[ans.dataset == 'III']
ds4 = ans[ans.dataset == 'IV']
```

```python
fig = plt.figure(figsize= (12, 10))
ax1 = plt.subplot(2, 2, 1)
ax2 = plt.subplot(2, 2, 2)
ax3 = plt.subplot(2, 2, 3)
ax4 = plt.subplot(2, 2, 4)

ax1.plot(ds1['x'], ds1['y'], 'o')
ax2.plot(ds2['x'], ds2['y'], 'o')
ax3.plot(ds3['x'], ds3['y'], 'o')
ax4.plot(ds4['x'], ds4['y'], 'o')

ax1.set_title('Dataset I')
ax2.set_title('Dataset II')
ax3.set_title('Dataset III')
ax4.set_title('Dataset IV')

fig.suptitle("Anscombe's Quartet")
fig.tight_layout()
plt.show()
```

#### code Refactoring

```python
fig, axes = plt.subplots(2, 2, figsize=(12,10), sharex=True, sharey=True)
for i, num in enumerate(['I', 'II', 'III', 'IV']):
    ds = ans[ans.dataset == num]
    ax = axes[i//2][i%2]
    ax.plot(ds.x, ds.y, 'o', markersize=10)
    ax.plot([3,20], [4.5,13], 'r-', lw=2)
    ax.set_title(f'Dataset {num}')

fig.suptitle("Anscombe's Quartet", fontsize=16)
fig.tight_layout()
fig.subplots_adjust(top=0.9)
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10), sharex=True, sharey=True)

for i, num in enumerate(['I', 'II', 'III', 'IV']):
    ds = ans[ans.dataset == num]
    ax = axes[i//2][i%2]
    ax.plot(ds.x, ds.y, 'o', markersize=10)
    ax.plot([3,20], [4.5, 13], 'r-', lw=2) 
    ax1.set_title(f"Dataset {num}")

fig.suptitle("Anscombe's Quartet", fontsize=16)
fig.tight_layout()
fig.subplots_adjust(top=0.9)
plt.show()
```

### Axis 객체와 축

```python
fig, ax0 = plt.subplots()
ax1 = ax0.twinx()

ax0.set_title("2개의 y축 한 figure에서 사용하기")
ax0.plot([10, 5, 2, 9, 7], 'r-', label="y0")
ax0.set_ylabel("y0")
ax0.grid(False)

ax1.plot([100, 200, 220, 180, 120], 'g:', label="y1")
ax1.set_ylabel("y1")
ax1.grid(False)

ax0.set_xlabel("공유되는 x축")

fig.legend(loc='upper right', bbox_to_anchor=(0.86, 0.89))
plt.show()
```
