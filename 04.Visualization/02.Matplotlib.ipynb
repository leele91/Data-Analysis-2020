### 08. Matplotlib 그래프 영역 채우기
- fill_between

```python
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

plt.plot(x, y)
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.fill_between(x[1:3], y[1:3], alpha=0.5) # alpha:투명도
# fill_between : 사용시 차트 내부 또는 두개의 선분 사이를 특정한 색으로 채울수 있음
# 네 점(x[1],y[1]),(x[2], y[2]),(x[2], 0),(x[3], 0)을 잇는 영역이 채워짐
# plt.fill_betweenx(y[2:4], x[2:4], color='pink', alpha=0.5)  

plt.axhline(0); plt.axhline(0) #axhline: 수직선 수평선
plt.grid(True)
plt.show()
```

```python
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

plt.plot(x, y)
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.fill_betweenx(y[2:4], x[2:4], color='pink', alpha=0.5) 
plt.axhline(0); plt.axhline(0)
plt.grid(True)
plt.show()
```

```python
x = np.arange(1.0, 4.0, 0.011)
y1 = x ** 2 #x의 제곱
y2 = x ** 2 / 2  #x의 제곱(** 2)

plt.plot(x, y1); plt.plot(x, y2)
plt.xlabel('X-Label');plt.ylabel('Y-Label')
plt.title('두 그래프 사이 영역 채우기')
plt.fill_between(x[100:200], y1[100:200], y2[100:200], color='lightgray', alpha=0.5)
# 네 점 (x[1], y[1]), (x[1], y[2]), (x[2], y[1]), (x[2], y[2]) 사이 영역
plt.grid(True)
plt.show()
```

```python
### 10. Matplotlib 그리드 설정하기
```

```python
a = np.arange(0, 2, 0.2)

plt.plot(a, a, 'bo')
plt.plot(a, a**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(a, a**3, color='springgreen', marker='^', markersize=9)
plt.grid(True)

plt.show()
```

```python
### 축 지정하기
# axis=y로 설정하면 가로 방향의 그리드만 표시됩니다.
# {‘both’, ‘x’, ‘y’} 중 선택할 수 있고 디폴트는 ‘both

a = np.arange(0, 2, 0.2)

plt.plot(a, a, 'bo')
plt.plot(a, a**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(a, a**3, color='springgreen', marker='^', markersize=9)
plt.grid(True, axis='y')
plt.title('축 지정하기')
plt.show()
```

```python
### 스타일 설정하기
#  which 파라미터를 ‘major’, ‘minor’, ‘both’ 등으로 사용하면 주눈금, 보조눈금에 각각 그리드를 표시
a = np.arange(0, 2, 0.2)

plt.plot(a, a, 'bo')
plt.plot(a, a**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(a, a**3, color='springgreen', marker='^', markersize=9)
plt.grid(True, axis='y', color='red', alpha=0.5, linestyle='--')
plt.title('그리드 스타일 설정하기')
plt.show()
```

### 13. Matplotlib 수직선/수평선 표시하기
- axhline/axvline
- hlines/vlines

```python
a = np.arange(0, 2, 0.2)

plt.plot(a, a, 'bo')
plt.plot(a, a**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(a, a**3, color='springgreen', marker='^', markersize=9)
plt.xticks(np.arange(0, 2, 0.2), labels=['Jan', '', 'Feb', '', 'Mar', '', 'May', '', 'June', '', 'July'])
plt.yticks(np.arange(0, 7), ('0', '1GB', '2GB', '3GB', '4GB', '5GB', '6GB'))

plt.axhline(1, 0, 0.55, color='gray', linestyle='--', linewidth='1')
plt.axvline(1, 0, 0.16, color='lightgray', linestyle=':', linewidth='2')

plt.axhline(5.83, 0, 0.95, color='gray', linestyle='--', linewidth='1')
plt.axvline(1.8, 0, 0.95, color='lightgray', linestyle=':', linewidth='2')

plt.show()
```

```python
a = np.arange(0, 2, 0.2)

plt.plot(a, a, 'bo')
plt.plot(a, a**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(a, a**3, color='springgreen', marker='^', markersize=9)
plt.xticks(np.arange(0, 2, 0.2), labels=['Jan', '', 'Feb', '', 'Mar', '', 'May', '', 'June', '', 'July'])
plt.yticks(np.arange(0, 7), ('0', '1GB', '2GB', '3GB', '4GB', '5GB', '6GB'))

plt.hlines(4, 1, 1.6, colors='pink', linewidth=3)
plt.vlines(1, 1, 4, colors='pink', linewidth=3)

plt.show()
```

### 14. Matplotlib 막대 그래프 그리기
- 범주형 데이터를 표현할때

```python
x = np.arange(3)
years = ['2017', '2018', '2019']
values = [100, 400, 900]

plt.bar(x, values) # 막대는 bar
# years는 x축에 표시될 연도이고, values는 막대 그래프의 높이로 표시될 y 값 
plt.xticks(x, years)
plt.show()
```

```python
### 스타일 꾸미기
```

```python
x = np.arange(3)
plt.bar(x, values, width=0.6, align='edge', color="springgreen",
        edgecolor="gray", linewidth=3, tick_label=years, log=True)
plt.title('스타일 꾸미기')
plt.show()
```

### 15. Matplotlib 수평 막대 그래프 그리기

```python
y = np.arange(3)
years = ['2017', '2018', '2019']
values = [100, 400, 900]

# height(두께)=0.8이 기본값
plt.barh(y, values, height=-0.8, align='edge', color="springgreen",
        edgecolor="gray", linewidth=3, tick_label=years, log=False)
plt.show()
```

```python
np.random.seed(19680801)    # Reproducible random state

N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N) # 칼라
area = (30 * np.random.rand(N))**2 # 크기

plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
```

```python
a = 2.0 * np.random.randn(10000) + 1.0
b = np.random.standard_normal(10000)
c = 20.0 * np.random.rand(5000) - 10.0

plt.hist(a, bins=100, density=True, alpha=0.7, histtype='step', label='a')
plt.hist(b, bins=50, density=True, alpha=0.5, histtype='stepfilled', label='b')
plt.hist(c, bins=100, density=True, alpha=0.9, histtype='step', label='c')
plt.legend() # 레이블 사용시 hist 뒤 옵션(lable)을 추가 하고 사용해야함
plt.show()
```

### 20. Matplotlib 파이 차트 그리기

```python
ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']

plt.pie(ratio, labels=labels, autopct='%.1f%%')
plt.show()
```

```python
plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False)
plt.show()
```

### 중심에서 벗어나는 정도 설정하기

```python
ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
explode = [0, 0.10, 0, 0.10]

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, explode=explode)
# explode는 부채꼴이 파이 차트 중심에서 벗어나는 정도를 설정 
plt.show()
```
