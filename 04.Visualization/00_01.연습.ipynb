```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
mpl.rc('font', family='Malgun Gothic') #family에 사용할 폰트를 넣음
# 유니코드에서  음수 부호설정
mpl.rc('axes', unicode_minus=False)
```

```python
plt.plot([0, 10, 15, 30],[5, 20, 25, 30])

plt.title('한글 제목')
plt.xlabel('X Axis') # x축 제목
plt.ylabel('Y Axis') # y축 제목
plt.grid(True)
plt.show()
```

```python
t = np.arange(0,10,2)

plt.plot(t, t, 'r--', t, t**2,'g^') # x축, y축,어떻게 표시하나
plt.grid(True)
plt.show()
```

```python
x = [0, 10, 15, 20]
y = [0, 10, 20, 30]

plt.plot(x, y)
plt.xlabel('x축')
plt.ylabel('y축')
plt.fill_between(x[2:4],y[2:4], alpha=0.5)
plt.show()
```

```python
x = [0, 10, 15, 20]
y = [0, 10, 20, 30]

plt.plot(x, y)
plt.xlabel('x축')
plt.ylabel('y축')
plt.fill_between(x[1:3],y[1:3], alpha=0.5)
plt.grid(True)
plt.show()
```

```python
x = [0, 10, 15, 20]
y = [0, 10, 20, 30]

plt.plot(x, y)
plt.xlabel('x축')
plt.ylabel('y축')
plt.fill_betweenx(y[2:4], x[2:4], alpha=0.5)
plt.grid(True)
plt.show()
```

```python
x = np.arange(1.0, 4.0, 0.011)
y1 = x ** 2 #x의 제곱
y2 = x ** 2 / 2  #x의 제곱(** 2)

plt.plot(x, y1); plt.plot(x, y2)
plt.xlabel('X-Label');plt.ylabel('Y-Label')
plt.title('두 그래프 사이 영역 채우기')

plt.fill_between(x[100:200], y1[100:200], y2[100:200], color='lightgray', alpha=0.5)
# 네 점 (x[1], y[1]), (x[1], y[2]), (x[2], y[1]), (x[2], y[2]) 사이 영역
plt.grid(True)
plt.show()
```
